### Clean:
1. remove empty columns form df
2. make the date the index of the df

```python
def clean_data(df):
    if 'Unnamed: 0' in df.columns:
        df = df.drop('Unnamed: 0', axis=1)
    else:
        pass
    if 'Unnamed: 0.1' in df.columns:    
        df = df.drop('Unnamed: 0.1', axis=1)
    else:
        pass
    if 'Unnamed: 0.1.1' in df.columns:
        df = df.drop('Unnamed: 0.1.1', axis=1)
    else:
        pass
    df = df.set_index(pd.to_datetime(df.Date))
    df = df.drop('Date', axis=1)
    df = df.sort_index()
    df = pd.get_dummies(df)
    return df
```

```python
apple = clean_data(apple)
cvs = clean_data(cvs)

wfc= clean_data(wfc)

abbvie = clean_data(abbvie)

multi_positions = pd.concat([apple, abbvie, wfc, cvs], axis=1)

multi_positions.shape
```

### Building Model with KPCA & SVR:
1. Extract X and y values
2. train, test, split
3. Grid Search
    - build pipeline
    - build grid
    - preform gridsearch
    - fit pipe


```python
#define the y's as the current day's trading signal (because we shift the data in the next step we have to drop the first day)
y = apple.iloc[1:, :].Target
#we shift the current days features to the previous days, so we can predict tomorrow's trading signal
apple= apple.iloc[:, 8:].shift()
#drop the NAN values created by the shift
apple = apple.dropna()
# define X's as the features without the targets
X = apple.drop('Target', axis=1)
```

```python
X.shape
```

```python
#because this is time-series data we must manually split the data 
X_train = X.iloc[0:900, :]
y_train = y.iloc[0:900]
X_test = X.iloc[901:, :]
y_test = y.iloc[901:]
```

```python
X_train.shape
```

```python
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('kpca', KernelPCA()),
    ('svr', SVR())
])
```

```python
kpca_params = {
    'kpca__n_components': np.arange(2, 11, 1),
    'kpca__kernel': ['rbf', 'sigmoid', 'cosine'],
    'kpca__gamma': np.linspace(.1, 1, 10),
    'svr__kernel':['sigmoid', 'rbf']
}
```

```python
pipe_gs = GridSearchCV(pipeline, param_grid= kpca_params, cv=5, n_jobs=-1)
```

```python
pipe_gs.fit(X_train, y_train)
```

```python
y_pred = pipe_gs.predict(X_test)
```

```python
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
# Explained variance score: 1 is perfect prediction
print('Explained Variance: %.2f' % r2_score(y_test, y_pred))
```
